```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
t_continuous = np.linspace(0,1,100)
sine_function = np.sin(20*np.pi*t_continuous)

sine_function.shape
```

```python
def plot_signal(x, y, legend_label=r"$y(t)$"3):
    plt.plot(x,y,label=legend_label)
    plt.xlabel(r"$t$")
    plt.ylabel(r"$y(t)$")
    plt.legend(loc='upper right')
    
    return

plot_signal(t_continuous, sine_function)
```
