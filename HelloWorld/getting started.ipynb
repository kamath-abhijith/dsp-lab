```python
import numpy as np
import IPython

from scipy.io import wavfile

from matplotlib import pyplot as plt
```

```python
t_continuous = np.linspace(0,1,100)
sine_function = np.sin(20*np.pi*t_continuous)
exp_function = np.exp(-2*t_continuous)

exp_function.shape
```

```python
def plot_signal(x, y, ax=None, legend_label=r"$y(t)$"):
    if ax is None:
        fig = plt.figure()
        ax = plt.gca()
    
    plt.plot(x, y, label=legend_label)
    plt.xlabel(r"$t$")
    plt.ylabel(r"$y(t)$")
    plt.legend(loc='upper right')
    
#     plt.xlim([0,1])
#     plt.ylim([-1,1])
#     plt.savefig()
    
    return

fig = plt.figure()
ax = plt.gca()
plot_signal(t_continuous, sine_function, ax=ax, legend_label=r"sine")
plot_signal(t_continuous, exp_function, ax=ax, legend_label=r"exp")
```

```python
ecg_data = np.loadtxt('ECG_Data.txt')

plot_signal(np.arange(ecg_data.shape[0]), ecg_data)

# 0:6883 MATLAB EQUIV.
```
